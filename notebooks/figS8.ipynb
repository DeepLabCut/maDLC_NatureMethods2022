## Reproduces Extended Data Figure 5 in Lauer et al., Nature Methods 2022

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='ticks')

%matplotlib inline
```

Fig S8d

```python
data_mouse = {'[0, 85]': 0.8581245526, '[0, 115]': 0.7998568361, '[15, 100]': 0.8191839656, '[15,120]': 0.8599856836, 'DLC':.998}
data_marmo = {'[0, 46]': -0.1395076202, '[5, 46]': -0.9594301751, '[10, 46]': -0.6161142492, '[15, 46]': -0.9635155778, 'DLC':.971}
names1 = list(data_mouse.keys())
values1 = list(data_mouse.values())
names2 = list(data_marmo.keys())
values2 = list(data_marmo.values())

sns.set(font_scale=1.5)
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
ax[0].scatter(names1, values1, s=60, c="#EEB98D")
ax[0].scatter(names1[-1], values1[-1], s=60, c="k")
ax[0].set_ylabel('MOTA')
ax[0].set_xlabel('[min, max] intensity range')
ax[1].scatter(names2, values2, c='#C6386C', s=60)
ax[1].scatter(names2[-1], values2[-1], c='k', s=60)
ax[0].axhline(0, ls=':', c='k')
ax[1].axhline(0, ls=':', c='k')
fig.savefig('figS8d.jpg', dpi=600)

plt.show()
```
