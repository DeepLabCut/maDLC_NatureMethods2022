## Reproduces Figure 4 in Lauer et al., Nature Methods 2022

```python
import matplotlib.pyplot as plt

import matplotlib as mpl

mpl.rcParams['axes.spines.left'] = True
#mpl.rcParams['axes.spines.right'] = False
#mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.bottom'] = True
```

```python
metrics = ['MOTA', 'False Positives', 'False Negatives', 'Switches', 'IDF1', 'IDP',
           'IDR', 'Recall', 'Precision', 'GT', 'MT', 'FM']
           


data = [[76.7, 0, 84, 9, 87.4,99,78.2,78.9,100,14,10,15],
        [85.2, 1, 46, 12, 92.2, 98, 87, 88.5, 100, 14, 11, 19],
        [98.8,0,18,0,99.4,100,98.8,98.8,100,2,2,12],
        [98.8,0,18,0,99.4,100,98.8,98.8,100,2,2,12]
       ]

for e in data:
    print (len(e))
count = 0
fig, axes = plt.subplots(3,4)
fig.set_size_inches(12, 10)
exps = ['baseline', 'TransID', 'baseline ', 'TransID ']
for i in range(len(metrics)):   
    for exp in exps:
        count+=1
        row = i//4
        column = i % 4
        baseline_data_fish = data[0][i]
        transID_data_fish = data[1][i]
        baseline_data_marmoset = data[2][i]
        transID_data_marmoset = data[3][i]
        all_ = [baseline_data_fish, transID_data_fish,
                baseline_data_marmoset, transID_data_marmoset]
        labels = exps
        #axes[row,column].set_axis_off()
        if min(all_)> 80:
            axes[row,column].set_ylim(80,100)
        
        axes[row,column].tick_params(axis= 'x',labelsize=16, rotation = 45)
        axes[row,column].tick_params(axis= 'y',labelsize=16)
        
        axes[row,column].set_title(metrics[row*4+column],fontsize = 18)
        if i <= 11:
            axes[row,column].set_axis_on()
            #ax.set_title(c)
        axes[row,column].bar(exps, all_ , color=['cornflowerblue','midnightblue','orchid','darkmagenta'])
       
        
#fig.patch.set_visible(False)
#plt.xticks(fontsize = 60)
fig.tight_layout(pad=1.0)
plt.show()
```

```python
# acc
metrics = ['train acc.' , 'test acc.', 'train acc. ', 'test acc. ', 'train acc.  ', 'test acc.  ']
# fish
           
data = [99.9,98.5,100,99,99.7, 97]
        

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
fig.set_size_inches(3.5, 2)
ax.set_ylim(80,100)                   
#exps = ['fish', 'marmoest']

ax.bar(metrics, data, color=['grey','black','grey','black','grey','black'])
ax.tick_params(axis= 'x',labelsize=16, rotation = 45)
ax.tick_params(axis= 'y',labelsize=16)
ax.set_title('triplet accuracy (local tracks)',fontsize = 16)
plt.show()
        
```
