## Reproduces Figure 1a in Lauer et al., Nature Methods 2022

```python
import os
from utils import plot_single_image_keypoints

COLORMAP = "Set3"
MARKERSIZE = 12
```

```python
folder = "../data/fig1"
params = {
    "mice": {
        "img": "mouse.jpg",
        "data": "mouse.h5",
    },
    "pups": {
        "img": "pups.jpg",
        "data": "pups.h5",
    },
    "marmosets": {
        "img": "marmo.jpg",
        "data": "marmo.h5",
    },
    "fish": {
        "img": "fish.jpg",
        "data": "fish.h5",
    }
}
for dataset, dict_ in params.items():
    output_name = "out_" + dataset.replace("h5", "jpg")
    plot_single_image_keypoints(
        os.path.join(folder, dict_["data"]),
        os.path.join(folder, dict_["img"]),
        MARKERSIZE,
        COLORMAP,
        output_name,
    )
```
